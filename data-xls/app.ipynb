```python
import pandas as pd
```

```python
def convert_egg_shell_cell(cell):
    if cell == "-":
        return 'Kosong'
    return cell

def convert_class_cell(cell):
    if cell == "Mammalia":
        return "Penyusu"
    return cell

df = pd.read_excel("files/animals.xlsx", "Animals", index=1, converters={
    "Egg Shell": convert_egg_shell_cell,
    "Class": convert_class_cell
})
```
